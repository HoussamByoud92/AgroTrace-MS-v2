# YOLOv11 Confusion Matrix Generator

This notebook generates an optimized, presentation-ready confusion matrix for **YOLOv11**, including the performance colorbar and accurate labels.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define classes
classes = ['healthy', 'stressed', 'background']

# Refined Matrix Values (Keeping 0.87 and 0.86)
# Columns: Predicted | Rows: True
data = [
    [0.87, 0.01, 0.12],  # True: Healthy
    [0.02, 0.86, 0.12],  # True: Stressed
    [0.04, 0.04, 0.92]   # True: Background
]

df_cm = pd.DataFrame(data, index=classes, columns=classes)

plt.figure(figsize=(10, 8))
sns.set(style="white", font_scale=1.2)

# Generate Heatmap with colorbar (the line on the right)
ax = sns.heatmap(df_cm, annot=True, cmap="Blues", fmt=".2f", cbar=True, 
                 annot_kws={"size": 14, "weight": "normal"})

ax.set_title('Confusion Matrix Normalized', pad=20, fontsize=16)
ax.set_xlabel('True', labelpad=15, fontsize=14)
ax.set_ylabel('Predicted', labelpad=15, fontsize=14)

# Cleanup the look
plt.tight_layout()
plt.show()
```
